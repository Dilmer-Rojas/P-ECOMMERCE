```python
# Importando librerias necesarias
import pandas as pd
```

```python
# Carga de datos
rutas = {
    'clientes': '../data/raw/clientes.csv',
    'categorias': '../data/raw/categorias.csv',
    'productos': '../data/raw/productos.csv',
    'ventas': '../data/raw/ventas.csv',
    'metodos_pago': '../data/raw/metodos_pago.csv'
}

data = {nombre: pd.read_csv(ruta) for nombre, ruta in rutas.items()}

clientes, categorias, productos, ventas, metodos_pago = data.values()
```
